**Sensor Function**

```python
def create_sensor(sensor_id, location, sensor_type, threshold=None):
    """
   Sensor data structure definition.
    """
    sensor = {
        "id": sensor_id,
        "location": location,
        "type": sensor_type,
        "threshold": threshold
    }
    return sensor
```

**Alert Function**

```python
def create_alert(severity, message, sensor_id, timestamp):
    """
    Creates an alert data structure.
    """
    alert = {
        "severity": severity,
        "message": message,
        "sensor_id": sensor_id,
        "timestamp": timestamp
    }
    return alert
```

**Creation of sensors**

```python
sensors = [
    create_sensor("S1", "Living Room", "motion"),
    create_sensor("S2", "Kitchen", "temperature", threshold=75),
    create_sensor("S3", "Front Door", "door"),
    create_sensor("S4", "Bedroom", "smoke")
]
```

```python
print(f"Initialized {len(sensors)} sensors")

for sensor in sensors:
    print(f"  - {sensor['id']}: {sensor['location']} ({sensor['type']})")
```
